```python
import pandas as pd
import numpy as np
```

```python
PATH = '../TrainingData/3000random_to_tag.xlsx'
```

```python
data = pd.read_excel(PATH)
data['year'] = data['year'].astype(str)
data['year'] = data['year'].apply(lambda x: x.split('-')[0])
data.head()
```

```python
data_resampled = data[data['Resample'] == 'Y']
data_not_resampled = data[data['Resample'] != 'Y']

data_resampled.head()
```

```python
complete_data = data_not_resampled.loc[:, ['id', 'file', 'year', 'newspaper', 'city', 'TAG', 'PositiveWords', 'NegativeWord', 'text']]
print(len(complete_data), len(data_resampled), len(complete_data) + len(data_resampled))
complete_data.head()
```

```python
new_data = {
    'id': [],
    'file': [],
    'year': [],
    'newspaper': [],
    'city':[],
    'TAG': [],
    'PositiveWords': [],
    'NegativeWord': [],
    'text': []
}
for i in range(len(data_resampled)):
    d = data_resampled.iloc[i]
    print(d)
    new_data['id'].append(d['id'])
    new_data['file'].append(d['file'])
    new_data['year'].append(d['year'])
    new_data['newspaper'].append(d['newspaper'])
    new_data['city'].append(d['city'])
    new_data['PositiveWords'].append(d['PositiveWords'])
    new_data['NegativeWord'].append(d['NegativeWord'])
    new_data['text'].append(d['Resample_1'])
    new_data['TAG'].append(d['TAG_1'])

    new_data['id'].append(d['id'])
    new_data['file'].append(d['file'])
    new_data['year'].append(d['year'])
    new_data['newspaper'].append(d['newspaper'])
    new_data['city'].append(d['city'])
    new_data['PositiveWords'].append(d['PositiveWords'])
    new_data['NegativeWord'].append(d['NegativeWord'])
    new_data['text'].append(d['Resample_2'])
    new_data['TAG'].append(d['TAG_2'])

data_new = pd.DataFrame(new_data)
```

```python
len(data_new)
```

```python
data_new = data_new.dropna(subset='TAG')
print(len(data_new))
data_new.head()
```

```python
result = pd.concat([complete_data, data_new])
print(len(result))
result.head()
```

```python
result = result.dropna(subset='TAG')
result['TAG'] = result['TAG'].astype(str).apply(lambda x: x.upper().strip())
```

```python
positive_words = result.PositiveWords.astype(str)
negative_words = result.NegativeWord.astype(str)
```

```python
pos_words = set()
for pos_word in positive_words:
    pos_word = pos_word.strip()
    if not (pos_word == '' or pos_word == 'nan'):
        words = pos_word.split('\n')
        for word in words:
            pos_words.add(word)

pos_words
```

```python
neg_words = set()
for neg_word in negative_words:
    neg_word = neg_word.strip()
    if not (neg_word == '' or neg_word == 'nan'):
        words = neg_word.split('\n')
        for word in words:
            neg_words.add(word)

neg_words
```

```python
with open('../Lexicon/Positive_words.txt', 'w') as file:
    for word in pos_words:
        file.write(word + '\n')

with open('../Lexicon/Negative_words.txt', 'w') as file:
    for word in neg_words:
        file.write(word + '\n')
```

```python
result = result.loc[:, ['id', 'file', 'year', 'newspaper', 'city', 'TAG', 'text']]
result = result[result['TAG'] != 'ILEGIBLE']
result.to_csv('../TrainingData/corpus_tagged.csv')
```
